```python
# beautifulsoup
from bs4 import BeautifulSoup
```

```python
with open('/content/drive/MyDrive/Colab Notebooks/scraped_data/data1.html', 'r') as f:
  html_content = f.read()
soup = BeautifulSoup(html_content,"lxml")
```

```python
soup
```

```python
soup.find("h1")
```

```python
all_h3= soup.find_all("h3")
for h3 in all_h3:
   print(h3.text)
```

```python
all_h3= soup.find_all("h3")
for h3 in all_h3:
  name = h3.get_text(strip=True)
  print(h3.find_parent("div")["class"])
```

```python
all_h3= soup.find_all("h3")
for h3 in all_h3:
  name = h3.get_text(strip=True)
  population =  h3.find_next("div").select("span.country-population")[0].get_text(strip=True)
  print(h3.find_next("div").select("span.country-population")[0].get_text(strip=True))
  print(h3.attrs)
```

```python
all_h3= soup.find_all("h3")
all_countries = []
for h3 in all_h3:
  name = h3.get_text(strip=True)
  population =  h3.find_next("div").select("span.country-population")[0].get_text(strip=True)
  print(h3.find_next("div").select("span.country-population")[0].get_text(strip=True))
  all_countries.append([name,population])
```
